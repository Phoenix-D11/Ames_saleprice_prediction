### Dealing with null values

```python
null = []
drop = []
for i in range(0,df.shape[1]):
    if 0 < df.iloc[:,i].isna().sum() <= df.shape[0]*(0.3):
        null.append(df.columns[i]) 
    elif df.iloc[:,i].isna().sum() > df.shape[0]*(0.3):
        drop.append(df.columns[i])
print(null, "\n")
print(drop)
```

```python
ty = df.select_dtypes("object")
for col in ty.columns.tolist():
    if col in null:
        print('{} column missing values: {}'.format(col, ty[col].isnull().sum()))
```

```python
des = []
for i in list(df):
    if i in list(null):
        if i not in list(ty):
            des.append(i)        
for col in des:          
    print('{} column missing values: {}'.format(col, df[col].isnull().sum()))
```
